### dependecies

```python
import requests
from bs4 import BeautifulSoup
import regex
```

### mining the data

```python
web_url = requests.get('https://www.worldometers.info/coronavirus/')
sp = BeautifulSoup(web_url.text, 'html.parser')

covid_data = sp.find_all('div', class_='maincounter-number')
print('Coronavirus Cases:', covid_data[0].text.strip())
print('Coronavirus Death:', covid_data[1].text.strip())
print('Coronavirus Recovered:', covid_data[2].text.strip())
```

```python
index = sp.find_all('script', {'type': 'text/javascript'})

result = sp.find(id='main_table_countries_today')
content = result.find_all('td')
clean_data = ''
for data in content:
  clean_data += data.text.strip() + '|'
  clean_data = clean_data.replace('+', '')
  clean_data = clean_data.replace('N/A', '0')

countries = []
locate = sp.find_all('a', class_='mt_a')
for country in locate:
  countries.append(country.text.strip())

full_data = []
for c in countries:
  pattern = c+'[,|\d]*'
  data = regex.search(pattern, clean_data, regex.IGNORECASE)
  if data != '':
    full_data.append(data.group())

full_data
```
